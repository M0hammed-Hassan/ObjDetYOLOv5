### Collecting The Data

```python
# Making the data dirs
os.makedirs('data/images')

# The data path
imgs_path = os.path.join('Data','images')

# The labels
labels = ['Awake','Drowsy']

# Capturing from the webcamera
cap = cv2.VideoCapture(0)

# Iterating over the labels
for label in labels:
    print(f'Collecting data for {label} label...')
    time.sleep(2)
    
    # Iterating over a range
    for i in range(25):
        print(f'Collecting image {i + 1} for {label} label')
        time.sleep(2)
        ret,frame = cap.read()
        img_path = os.path.join(imgs_path,label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(img_path,frame)
        cv2.imshow('Collecting....',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

### Real Time Object Detection

```python
# Loading the model
model = torch.hub.load('ultralytics/yolov5','custom',path = 'yolov5/runs/train/exp2/weights/last.pt',force_reload=True)
```

```python
model
```

```python
# Capture a vido
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame = cap.read()
    pred = model(frame)
    pred = pred.render()[0]
    cv2.imshow('Real Time Detection',pred)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
